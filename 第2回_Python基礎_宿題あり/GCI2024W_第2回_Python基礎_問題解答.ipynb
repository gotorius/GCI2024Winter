(3-3) 要素がすべて整数値であって、長さが1以上のリストが変数xに格納されているとします。xの要素の中で最も大きい要素を返すプログラムを作成してください。ただし、max関数などは使用せず、ループ（for文での繰り返し）で実現してください。

```python
# 入力例
x = [1, 5, 4, 3]
```

```python
# 最大値を変数maxに格納する
max = x[0] # maxの初期値は配列の先頭にする
for i in x:
    if max < i:
        max = i # maxをiに入れ替える
print(max)
```

## 4.2 練習問題

(4-1) 長さが1以上の文字列を引数`x`とし、`x`の最後の1文字を返す関数`f`を定義してください。

```python
# 入力例
x = 'python'
```

```python
def f(x):
    length = len(x) # 文字列の長さ
    return x[length - 1] # 0から始まるので、lengthより１つ小さいインデックスを指定
f(x)
```

```python
# 別解
def f(x):
    return x[-1]
f(x)
```

(4-2) 辞書を引数`x`とし、`x`に格納されているキーの数を返す関数`f`を定義してください。

```python
# 入力例
x = {"a": 1, "b": 2, "c": 3, 'd':4, 'e':5}
```

```python
def f(x):
    num_key = x.keys() # keys()は、辞書のキーを取得する関数
    return len(num_key) # num_keyの長さがキーの数となる
f(x)
```

(4-3) リストを引数`x`とし、`x`の要素を逆順に並べたリストを返す関数`f`を定義してください。ただし、`reversed`関数などは使用せず、ループやリストのスライスで実現してください。

```python
# 入力例
x = [3, 9, 7, 1, 0]
```

```python
def f(x):
    reverse = [] # 結果を入れる変数(空箱)を用意する
    length = len(x) # 繰り返し回数
    for i in range(1,length + 1):
        reverse.append(x[-i]) # 後ろに追加していく
    return reverse

f(x)
```
